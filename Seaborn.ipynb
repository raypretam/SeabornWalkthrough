## Introduction to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
tips_data = sns.load_dataset('tips')
flights_data = sns.load_dataset('flights')
tips_data.head()
```

```python
flights_data.head()
```

```python
sns.distplot(tips_data['total_bill']) #kde=kernel density estimation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips_data,kind='reg')
```

```python
sns.pairplot(tips_data,hue='sex')
```

## Categorical Plots

```python
sns.barplot(x='sex',y='tip',data=tips_data)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips_data,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_data,hue='smoker',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_data,split=True)
sns.swarmplot(x='day',y='total_bill',data=tips_data,color='black')
```

## Matrix Plots

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data (later in the machine learning section we will learn how to formally cluster data).

```python
flight = flights_data.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(flight,linecolor='white',linewidths=1)
```

```python
sns.clustermap(flight)
```

## Grids

```python
iris_data = sns.load_dataset('iris')
#iris_data.set_index('species',inplace=True)
iris_data.head()
```

```python
g = sns.PairGrid(iris_data)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris_data,hue='species',palette='rainbow')
```

## Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the lmplot() function for now.

lmplot allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
sns.lmplot(x='total_bill',y='tip',data=tips_data,hue='sex',markers=['o','v'],
           scatter_kws={'s':10}) #scatter_kws is calling matplotlib under the hood to
                                 #render functionality to seaborn.
```
